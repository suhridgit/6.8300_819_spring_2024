# Colab-to-pdf converter

# Warning: Pdf conversions sometimes cut off important information; that's probably why Google cut off pdf conversion support, so be sure to check your conversion that it retains your answers.

Runtime maybe 5 minutes

```python
import os
from google.colab import drive, files
from pathlib import Path
!pip install pyppeteer
drive.mount('/content/drive')

%pwd
#TODO: edit 'matcher' to find your .ipynb filename
#TODO: OR avoid brute force file search approach and fill in mynotebook_path below
matcher= "pset5.ipynb"

## Search your entire Drive for that filename! Brute force
potential_matches = [item for item in Path('/content/drive/MyDrive').rglob(matcher)]
for item in potential_matches:
  print("A match: ", str(item))
if len(potential_matches) < 1:
  print("No matches found.")
elif len(potential_matches) > 2:
  print("Duplicate files, unclear which.")

mynotebook_path = potential_matches[0]

# Now we convert to pdf and download
mynotebook_path = f'"{str(mynotebook_path)}"' ## Enclosing in "" allows for spaces in filenames to be interpreted correctly in the command line
mynotebook_pdfname = "/content/pset5_converted"
!jupyter nbconvert --to webpdf --disable-chromium-sandbox --WebPDFExporter.disable_sandbox=True --allow-chromium-download --output $mynotebook_pdfname $mynotebook_path

%ls
files.download(mynotebook_pdfname+".pdf")
# Kindly rename your pdf file to your kerboros id
```
